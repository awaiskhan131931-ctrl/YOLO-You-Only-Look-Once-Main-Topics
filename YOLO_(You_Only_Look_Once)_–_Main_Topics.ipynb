**Object detection**
Object detection is computer vision technique that identify and locates object within image or vedios.

Its does to things simultineusly:
* **Classification:** reconize what object are present.
(e.g: "car","person","dog")

* **localization:** determine where the object typically by drawing a bounding box around each object.

**How it works (modern approach):**
* An image pass through a neural network (usually CCN or Transfermer)
* The model predict the bounding box coordinate + class label + confidance scores
* A threshold filter out low-confidence detections

**Common applications:**

* Self-driving cars (detecting pedestrians, signs, vehicles)
* Security cameras and surveillance
* Medical imaging (detecting tumors, anomalies)
* Retail (inventory tracking, checkout systems)
* Augmented reality
* Face detection

**Classification vs Detection**
| Aspect | Classification | Detection |
|---------|---------------|------------|
| Question Answered | What is in the image? | What is it, and where is it? |
| Output | A label (+ confidence score) | Bounding boxes + labels + scores |
| Multiple Objects | Single prediction for the whole image | Can detect many objects at once |
| Localization | No | Yes |
| Complexity | Simpler | More complex |
| Speed | Faster | Slower |
| Example Models | ResNet, VGG, EfficientNet | YOLO, Faster R-CNN, SSD |

**Simple Visual Analogy**
**Classification:**

"This image contains a dog."

**Detection:**

"There is a dog at coordinates [120, 50, 300, 400] and a cat at [400, 80, 580, 350]."

##**YOLO (You Only Look Once)**

YOLO is a real-time object detection algorethem that object in image with a single pass through a neural networks -- hence the name.
** The core idea **
Traditional detector (like CNN) scan image multiple times in different regions.
**YOLO does all in one short **
* Divide image into a **S*S grid**
* Each grid cell predict bounding boxes + confidence score + class propabalities
* Output all detection at once

**YOLO Architecture:**
YOLO **(you look only once)** is a real time object detection algorethem that processes an image in a single forword pass through a neural network -- unlike traditional method that scan the image multiple times.

**the core idea:**
Divide the image into a grid and for each cell predict bounding boxes and class probability simulteniusly. this makes its extremely fast.

**The Three Main Components:**
1.  **backbone:**
The backbone is the feature extractor it processes the raw input image and learn to reperesent it as rich feature maps.
* Take the input image (e.g: 640x640 pixels)
* Applies deep convolutional layers to extract feature at multiple scales.
* Early layers detect low-level features(edges,texture,colors)
* Deeper layers detect high-level features (shapes, object parts)
* common bacbone use in YOLO variants:
  * Darknet-19 / Darknet-53 (YOLOv2, YOLOv3)
  * CSPDarknet (YOLOv4, YOLOv5)
  * EfficientRep (YOLOv6)
  * GELAN (YOLOv9)

Think of the backbone as the **"eyes"** of the model — it sees and encodes the image.

2.  **NECK:** The neck is the feature agregation layers ---its set between the backbone and head, and its job is to combine feature from different scale.
